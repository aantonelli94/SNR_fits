# Exploration of the dataset

## The dataset

We are dealing with a synthetic dataset (created separately as described in the paper), with the following features:
- $m_1\qquad $  (mass of the BH's primary), 
- $m_2\qquad$ (mass of the secondary), 
- $z\qquad$ (redshift) and 
- $\rho_{\rm avg}\qquad$ (SNR averaged over choices for all other parameters).

It's good practice to check if there are some null values. Fortunately there aren't in this dataset.

```python
df = pd.read_pickle("./dataset_SNRavg.pkl") 
df.columns = ['m1', 'm2', 'z', 'SNR']
df.head()
```

## Data wrangling

To help the fitting procedure we choose variables that are easier to work with. The first two columns are the component masses $m_1\, $ and $m_2 $, which can vary by orders of magnitude. As commonly done in GW astrophysics, we instead opt for sampling through the *log* of the total mass and the *symmetric mass ratio*.
These are defined as:

$$\log_{10}(M_\text{tot})=\log_{10}(m_1 + m_2), \quad \nu = (m_1 m_2)/(m_1 + m_2)^2$$

As these are of order unity, they lead to less variability in the features the fit needs to capture, and overall to a better performance.

```python
# First sum up the m1 and m2 columns into a single total mass and the ratio into a mass ratio column
df['Mtot'] = df.loc[:,['m1','m2']].sum(axis=1)
df['massratio'] =  (df['m1']*df['m2'])/(df['Mtot']**2) # mass ratio: df['m1']/df['m2']


# Then take the log10 of the SNR and total mass for easier fits.
df['logSNR'] = np.log10(df['SNR'])
df['logMtot'] = np.log10(df['Mtot'])


# Then we define the new dataframe to contain only total mass, redshift and log of the SNR.
df=df[['logMtot','massratio','z','logSNR']]
```

```python
df.head()
```

The error shown above is because there are some values that are infinity. Owing to how the dataset has been calculated (see paper), this is presumably because they are high redshift/high total mass sources. Indeed, such events sift out of band, which yields a zero SNR by default. 

Since infinities create problems for fits, we remove them.

```python
# Isolate a dataframe corresponding to SNR = 0
df_infsubset = df.loc[df['logSNR'] == -np.inf]

logMtot_infsubset = df_infsubset.logMtot
z_infsubset = df_infsubset.z
```

The population of SNR=0 binaries is formed of either high mass sources (which would be out of band), or high redshift sources, which would be too faint to be detected.

```python
# Plot this in a (Mtot,z) plane

plt.scatter(z_infsubset, logMtot_infsubset)
plt.ylabel('Mtot')
plt.xlabel('z')
```

In fact, we could remove also the values that have very low SNRs. These are not that interesting for the fit. We just need to make sure we give enough information around the threshold. 

We decide to fit only values with SNR bigger than 0.1: there's plenty of information in the remaining dataset, since usual thresholds start at SNR=8.

```python
# Remove the values with very low SNRs
rho_min = 0.1
df = df[df.logSNR > np.log10(rho_min)]

print('Minimum SNR', min(10**df['logSNR'].values))
```

## Checking correlations and multicollinearity

There appears no strong correlation or multicollinearity.

The interesting patterns appear between the SNR and total mass. That's what we hope to fit.

# Model

## Model fitting

We begin the fit by splitting the dataframe into training and test set, reserving 30$\%$ of the data for the test data.

```python
from sklearn.model_selection import train_test_split

#require a certain SNR threshold in the data frame to avoid numerical issues.
SNR_mask = df.logSNR > 1e-3
df = df[SNR_mask]

train, test = train_test_split(df, random_state = 2, test_size=0.3)
```

We split the training and test sets into $(X,y) $ variables, where $X $ are the independent variables and $y $ the dependent variable -- e.g., the SNR.

```python
variables = ['logMtot','massratio','z'] 

X_train = train[variables]
X_test = test[variables]
y_train = train.logSNR
y_test = test.logSNR
```

Finally we perform the polynomial regression fit -- to 5th order here -- using $\texttt{scikit-learn} $ in the lines below.

```python
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures

# Call the fit.
poly = PolynomialFeatures(degree=5)

# Transform the data so that it can be accepted by a linear model fit.
X_train_new = poly.fit_transform(X_train)
X_test_new = poly.fit_transform(X_test)

# Fit the model as if it was linear regression.
model = linear_model.LinearRegression()
polyfit = model.fit(X_train_new, y_train)

y_pred = polyfit.predict(X_test_new)
```

We get some metrics from statsmodels' ordinary least square regression.

```python
import statsmodels.api as sm


X = sm.add_constant(X_train_new.ravel())
results = sm.OLS(y_train,X_train_new).fit()
results.summary()  
```

## Model validation

In linear regression, dependent and independent variable are described by a linear relationship. This can be checked with a scatter plot between actual and predicted values.

```python
f = plt.figure(figsize=(14,5))
ax = f.add_subplot(121)
sns.scatterplot(y_test,y_pred,ax=ax,color='r')
```

The residual error should be normally distributed, and its mean close to zero.

```python
f = plt.figure(figsize=(14,5))
ax = f.add_subplot(121)

delta_y  = (10**y_pred- 10**y_test)/ 10**y_test
sns.distplot(delta_y,ax=ax,color='b')
ax.axvline(delta_y.mean(),color='k',linestyle='--')

ax.set_ylabel('count')
ax.set_xlabel(r'Relative difference in true SNR')
```

It's interesting to see if the same residual around zero is achieved for SNRs $8 < \rho < 15 $. This range is where one would place a threshold when using the above fit in a classification scheme. 

We check below that the same relative error is achieved for "low" SNR sources.

```python
f = plt.figure(figsize=(14,5))
ax = f.add_subplot(121)

mask_lowSNR = np.logical_and(y_test.values > np.log10(8),y_test.values < np.log10(15))
dy_lowSNR =  ((10**y_pred[mask_lowSNR]- 10**y_test[mask_lowSNR])/ 10**y_test[mask_lowSNR]).values
dy = ((10**y_pred- 10**y_test)/ 10**y_test).values 

sns.distplot(dy,ax=ax,color='b', label='All SNRs')
sns.distplot(dy_lowSNR,ax=ax,color='orange',label='SNR between 8 and 15')

ax.set_ylabel('count')
ax.set_xlabel(r'Relative difference in true SNR')
ax.legend(loc='best')
```

Finally, we find some metrics for the maximum outlier to get a grasp where the fit is failing.

```python
count = -1

index = np.where(dy == np.sort(dy)[count])[0][0]#np.argmax(np.abs(dy))
print('dy=',np.sort(dy)[count])

ypred_maxoutlier = y_pred[index]
ytest_maxoutlier = y_test.values[index]

print('Max outlier located at', (ypred_maxoutlier - ytest_maxoutlier)/ytest_maxoutlier)
print('SNR predicted',10**ypred_maxoutlier)
print('SNR test',10**ytest_maxoutlier)
print('SNR predicted/ SNR test',10**ypred_maxoutlier/10**ytest_maxoutlier)
```

Where does the point lie?

```python
print('Total mass at outlier:',df['logMtot'].values[index])
print('mass ratio at outlier:',df['massratio'].values[index])
print('Redshift at outlier:',df['z'].values[index])
```

# Checking the fit against externally-generated mock populations

Below is an example function showing how the fit could be used.

```python
def SNR_fit(Mtot, eta, z):
    
    """
    Function to return the SNR from a given 
    total mass, mass ratio, and redshift.
    """
    
    logMtot = np.log10(Mtot)
    input_ = [[logMtot, eta, z]]
    X_temp = poly.fit_transform(pd.DataFrame(input_))
    
    return 10**polyfit.predict(X_temp)[0]
```

----

We first check the results against mock SNR calculations from a random population. (This has been done offline using the $\texttt{gwbench}$ software, and saved into a pickle file.) The file contains random masses and redshift across the same range used for the fit, and the associated SNR that we should be able to broadly predict.

Furthermore, we transform the component masses of the validation dataset into log of the total mass and symmetric mass ratio.

```python
# Here is a validation dataset generated with gwbench.

df_validation = pd.read_pickle("./validation_N20_dataset_SNRavg.pkl") 
# df_validation = pd.read_pickle("./validation_dataset_SNRavg.pkl") 
df_validation.columns = ['m1', 'm2', 'z', 'SNR']
df_validation
```

```python
# transform m1 and m2 to logMtot and mass ratio

df_validation['Mtot'] = df_validation.loc[:,['m1','m2']].sum(axis=1)
df_validation['massratio'] =  (df_validation['m1']*df_validation['m2'])/(df_validation['Mtot']**2)
df_validation['logSNR'] = np.log10(df_validation['SNR'])
df_validation['logMtot'] = np.log10(df_validation['Mtot'])
df_validation=df_validation[['logMtot','massratio','z','logSNR']]

df_validation.head()
```

Here we isolate the values predicted through our fitting function and the ones obtained offline.

```python
SNR_gwbench = (10**df_validation['logSNR'].values)
SNR_predicted = [SNR_fit(10**(df_validation['logMtot'][i]),
        df_validation['massratio'][i], 
        df_validation['z'][i]) for i in range(0,len(df_validation['logSNR'].values))];
```

We then choose an SNR threshold and plot the scatter between mock and predicted populations,

```python
# Money plot

SNR_threshold = 12
plt.scatter(SNR_gwbench, SNR_predicted, marker='^',s=20)
plt.ylim(0,2*SNR_threshold)
plt.xlim(0,2*SNR_threshold)

plt.axvline(SNR_threshold, color='red')
plt.axhline(SNR_threshold, color='red')


plt.ylabel('Predicted SNR')
plt.xlabel('True SNR from gwbench')
```

This plot provides a great deal of information, and it shows how the fit can be used as a classification fit. Whenever both predicted and mock SNRs are below the threshold, we are in the upper right panel. In this case, the fit is successful in predicted a source that can be seen with the network specification chosen. When both are negative, we are in the lower left panel. In this case the fit is successful in making sure these events would not be seen. 

The interesting parts are the diagonal panels. In the upper left corner, we enter the realm of *false positives*; in the lower right corner, the realm of the *false negatives*. 

We study both cases in detail below.

-----

**False negatives** (SNRs predicted below the threshold, but that should be above)

-----

How many are they?

```python
cond = df_validation['logSNR'] < np.log10(SNR_threshold) # Keep mock SNRs above the threshold.
df_validation_masked = df_validation.mask(cond=cond).dropna()
nsamples = len(df_validation_masked)
print('Total number of SNRs predicted above the SNR threshold:',nsamples)
```

Scatter plot for the false negatives

```python
# False negatives plot.

SNR_gwbench_masked_neg = (10**df_validation_masked['logSNR'].values)
SNR_predicted_masked_neg = [SNR_fit(10**(df_validation_masked['logMtot'].values[i]),
        df_validation_masked['massratio'].values[i], 
        df_validation_masked['z'].values[i]) for i in range(0,len(df_validation_masked['logSNR'].values))];

plt.scatter(SNR_gwbench_masked_neg, SNR_predicted_masked_neg,  marker='*',s=20)

plt.ylim(0,2*SNR_threshold)
plt.xlim(0,2*SNR_threshold)
plt.axvline(SNR_threshold, color='red')
plt.axhline(SNR_threshold, color='red')


plt.ylabel('Predicted SNR')
plt.xlabel('True SNR from gwbench')
```

```python
# True is equal to 1.
n_falseneg = sum([SNR < SNR_threshold for SNR in SNR_predicted_masked_neg])

print('Number of false negatives:', n_falseneg)
print(round(n_falseneg/nsamples * 100,2),' % are predicted below the SNR threshold, but should be above.')
```

```python
SNR_mask_predicted = [SNR < SNR_threshold for SNR in SNR_predicted_masked_neg]
false_negatives = np.array(SNR_predicted_masked_neg)[SNR_mask_predicted]

print('The false negatives are:', false_negatives)
```

-----

**False positives** (SNRs predicted above the threshold, but that should be below)

-----

How many are they?

```python
cond = df_validation['logSNR'] > np.log10(SNR_threshold) # Keep mock SNRs below the threshold.
df_validation_masked = df_validation.mask(cond=cond).dropna()
nsamples = len(df_validation_masked)
print('Total number of SNRs predicted above the SNR threshold:',nsamples)
```

```python
SNR_gwbench_masked_pos = (10**df_validation_masked['logSNR'].values)
SNR_predicted_masked_pos = [SNR_fit(10**(df_validation_masked['logMtot'].values[i]),
        df_validation_masked['massratio'].values[i], 
        df_validation_masked['z'].values[i]) for i in range(0,len(df_validation_masked['logSNR'].values))];
```

```python
# False positives plot.

plt.scatter(SNR_gwbench_masked_pos, SNR_predicted_masked_pos,  marker='*',s=20)

plt.ylim(0,2*SNR_threshold)
plt.xlim(0,2*SNR_threshold)
plt.axvline(SNR_threshold, color='red')
plt.axhline(SNR_threshold, color='red')


plt.ylabel('Predicted SNR')
plt.xlabel('True SNR from gwbench')
```

```python
# True is equal to 1.
n_falsepositive = sum([SNR > SNR_threshold for SNR in SNR_predicted_masked_pos])

print('Number of false negatives:', n_falsepositive)
print(round(n_falsepositive/nsamples * 100,2),' % are predicted above the SNR threshold, but should be below.')
```

```python
SNR_mask_predicted = [SNR > SNR_threshold for SNR in SNR_predicted_masked_pos]
false_positives = np.array(SNR_predicted_masked_pos)[SNR_mask_predicted]

print('The false positives are:', false_positives)
```

---

**Conclusions**: overall, the fit is an accurate tool to classify whether a given event is above or below a threshold, with a percentage of $0.98\%$ false negatives and only $0.58\%$ false positives. The fit can be used as a reliable tool in conjunction with astrophysical-population codes.
